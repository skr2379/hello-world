```python
import tkinter

#Creating a new window and configurations
window = tkinter.Tk()
window.title("Widget Examples")
window.minsize(width=500, height=500)
window.config(padx=10, pady=10)

#Labels
label = tkinter.Label(text="This is old text")
label.grid(column=0, row=0)

label1 = tkinter.Label(text="text for checkbox status")
label1.grid(column=0, row=4)

label2 = tkinter.Label(text="text for radio button status")
label2.grid(column=0, row=5)

label3 = tkinter.Label(text="text for listbox selection")
label3.grid(column=0, row=8)

label4 = tkinter.Label(text="text for textbox selection")
label4.grid(column=0, row=11)

#Buttons
def action():
    label.config(text="This is new text")
    print("Do something")
def action1():
    entry_text = entry.get()
    label.config(text=entry_text)
def action2():
    spin_value =100-int(spinbox.get())
    label.config(text = spin_value)
def action3(value):
    value =entry.get()
    label.config(text = value)
def action4():
    label1.config(text = checked_state.get())
def action5():
    label2.config(text = f"radiob button {radio_state.get()} selected")
def listbox_used(event):
    list_value = listbox.get(listbox.curselection())
    label3.config(text = f"{list_value} selected from listbox")
def action6():
    
    label4.config(text = text.get("1.0",tkinter.END))

#calls action() when pressed
button = tkinter.Button(text="Click Me", command=action)
button.grid(column=0, row=1)
button.config(padx=5, pady=5)

button1 = tkinter.Button(text="entry_button", command=action1)
button1.grid(column=1, row=1)
button1.config(padx=5, pady=5)

button2 = tkinter.Button(text="textbox_button", command=action6)
button2.grid(column=3, row=1)
button2.config(padx=5, pady=5)

#Entries
entry = tkinter.Entry(width=30)
#Add some text to begin with
entry.insert(0,string="write something")
#Gets text in entry
print(entry.get())

entry.grid(column=1, row=0)

#spinbox
spinbox = tkinter.Spinbox(from_=0, to=100, width=5, command=action2)
spinbox.grid(column=2,row=2)

#scale
scale = tkinter.Scale(from_=0, to=100, width=5, command=action3)
scale.grid(column=3,row=2)

#checkbutton
checked_state = tkinter.IntVar()
checkbutton = tkinter.Checkbutton(text="is on", variable = checked_state, command = action4)
checkbutton.grid(column =3, row=3)
print(checked_state.get())

#radiobutton
radio_state = tkinter.IntVar()
radiobutton1 = tkinter.Radiobutton(text="option1",value=1, variable = radio_state, command = action5)
radiobutton1.grid(column =3, row=4)
radiobutton2 = tkinter.Radiobutton(text="option2",value=2, variable = radio_state, command = action5)
radiobutton2.grid(column =3, row=5)
print(radio_state.get())

#listbox
listbox = tkinter.Listbox(height=3)
fruits = ["apple","pear","orange","mango","peach","avacado","guava"]
for items in fruits:
    listbox.insert(fruits.index(items),items)
listbox.bind("<<ListboxSelect>>",listbox_used)
listbox.grid(column=0, row=6)

text = tkinter.Text(height=5, width=30)
text.focus()
text.insert(tkinter.END,"hello")
text.grid(column=5, row=5)
print(text.get("1.0",tkinter.END))

window.mainloop()
```
